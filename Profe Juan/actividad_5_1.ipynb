### Calcular una columna que indique el tiempo

```python
my_data["time"] = my_data.index*3
my_data
```

### Mover la columna a la primera posiciÃ³n

```python
column_to_move = "time"
columns = [column_to_move] + [col for col in my_data.columns if col != column_to_move]
my_data = my_data[columns]
my_data
```

### Ejemplo 1: Graficar el valor de las variables vs el tiempo

```python
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(20, 10))

axs[0,0].plot(my_data['time'], my_data['Veloc'])
axs[0, 0].xaxis.set_minor_locator(ticker.MultipleLocator(250))

axs[0,1].plot(my_data['time'], my_data['Acele'])

my_data['Des_x'].plot(ax=axs[1,0]) #Cuidado los valores de x serÃ¡n los del Ã­ndice

my_data.plot(x = 'time', y='Des_y', ax=axs[1,1], color='red') 


axs[1,1].set_title('GrÃ¡fica desplazamiento en y')
axs[1,1].set_xlabel('Tiempo')
axs[1,1].set_ylabel('Desplazamiento en y')
plt.show()
```

### Ejemplo 2: Histogramas, densidad y manejo del zoom

```python
min_velocidad = my_data["Veloc"].min()
max_velocidad = my_data["Veloc"].max()
print(f"MÃ­nio velocidad: {min_velocidad} Maximo velocidad: {max_velocidad} ")
```

```python
n = 10
intervalos = np.linspace(min_velocidad, max_velocidad, n + 1)

# Usa pd.cut con `labels` como el rango de intervalos en formato de string
my_data['Intervalo'] = pd.cut(
    my_data['Veloc'],
    bins=intervalos,
    labels=[f"[{intervalos[i]:.2f}, {intervalos[i+1]:.2f})" for i in range(n)],
    include_lowest=True
)

my_data[['Veloc', 'Intervalo']]
```

```python
columns_to_graph = ['Veloc', 'Acele', 'Des_x', 'Des_y']


intervals_cols = []
for column in columns_to_graph:
    min_val = my_data[column].min()
    max_val = my_data[column].max()
    intervalos = np.linspace(min_val, max_val, n + 1)

    # Usa pd.cut con `labels` como el rango de intervalos en formato de string
    my_data['Intervalo '+column] = pd.cut(
        my_data[column],
        bins=intervalos,
        labels=[f"[{intervalos[i]:.2f}, {intervalos[i+1]:.2f})" for i in range(n)],
        include_lowest=True
    )

    intervals_cols.append('Intervalo '+column)

my_data[columns_to_graph+intervals_cols]
```

```python
columns_to_graph = ['Veloc', 'Acele', 'Des_x', 'Des_y']

fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(20, 10))

# Cuenta la frecuencia de cada intervalo
frecuencias = my_data['Intervalo Veloc'].value_counts(sort=False)
axs[0, 0].bar(frecuencias.index.astype(str), frecuencias.values)
axs[0, 0].set_xticklabels(frecuencias.index.astype(str), rotation=45, ha="right")

#axs[0,1].hist(my_data['Veloc'], alpha=0.5, density=True, edgecolor ='black', bins=10)
axs[0,1].hist(my_data['Veloc'], alpha=0.5, edgecolor ='black', bins=10)

my_data['Acele'].plot.density(ax=axs[1,0])
# axs[1,0].set_xlim(0, 60)

my_data['Acele'].plot.density(ax=axs[1,1], color='red',)
axs[1,1].grid(linestyle="-", linewidth=0.5, color='green') #Debemos ponerlo despuÃ©s de la creaciÃ³n de la grÃ¡fica
axs[1,1].xaxis.set_major_locator(MultipleLocator(0.04))
axs[1,1].xaxis.set_minor_locator(AutoMinorLocator(5))
axs[1,1].set_xlim(0, 0.5)
axs[1,1].set_ylim(0, 50)
axs[1,1].set_title('GrÃ¡fica alejada')
axs[1,1].set_xlabel('Intervalo')
axs[1,1].set_ylabel('Frecuencia')
plt.show()
```

### Ejemplo 3 CÃ¡lculo de la distancia recorrida

```python
def distance(x, y, z):
    return math.sqrt(x**2 + y**2 + z**2)

my_data['Distancia'] = my_data.apply(lambda row: distance(row['Des_x'], row['Des_y'], row['Des_z']), axis=1)

my_data['Distancia'] 
```

```python
fig, axs = plt.subplots(nrows=3, ncols=1, figsize=(20, 10))
axs[0].plot(my_data['time'], my_data['Distancia'], color='blue')

plt.show()
```
